```python
x = c.iloc[:,1:].values
y = c.iloc[:,0:1].values
```

```python
y.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("oh",OneHotEncoder(),[0])],remainder = "passthrough")
y = ct.fit_transform(y).todense()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
xtn,xts,ytn,yts = train_test_split(x,y,test_size = 0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtn = sc.fit_transform(xtn)
xts = sc.fit_transform(xts)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
 ann = Sequential()
```

```python
ann.add(Dense(6,kernel_initializer = 'random_uniform',activation = 'relu'))
```

```python
ann.add(Dense(12,kernel_initializer = 'random_uniform',activation = 'relu'))
```

```python
ann.add(Dense(7,kernel_initializer = 'random_uniform',activation = 'softmax'))
```

```python
ann.compile('adam','categorical_crossentropy',metrics = ['accuracy'])
```

```python
ann.fit(xtn,ytn,batch_size = 32,epochs = 700)
```

```python
yp = ann.predict(xts)
```
